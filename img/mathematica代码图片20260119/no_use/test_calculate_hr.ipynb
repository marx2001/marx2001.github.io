```python
from pythtb import W90
import matplotlib.pyplot as plt
import numpy as np

# 1️⃣ 手动设置晶格常数与晶格向量
# 假设 2D 方格晶格，晶格常数 a = 3.9 Å（你可以换成你材料的真实数值）
a = 4.16
lat_vecs = [[a, 0.0], [0.0, a]]

# 2️⃣ 手动设置原子轨道（单位为晶格单位），例如两个轨道在原胞内：
orb_vecs = [[0.0, 0.0], [0.5, 0.5]]  # 自定义原胞中轨道相对位置

# 3️⃣ 使用 W90 构造 tight-binding 模型
nso = W90(hr_file="wannier90_hr.dat", lat=lat_vecs, orb=orb_vecs)

# 4️⃣ 设置高对称点路径
k_path = [
    [0.0, 0.0],     # Γ
    [0.5, 0.0],     # X
    [0.5, 0.0],     # X
    [0.5, 0.5],     # M
    [0.5, 0.5],     # M
    [0.0, 0.5],     # Y
    [0.0, 0.5],     # Y
    [0.0, 0.0]      # Γ
]
k_labels = [r'$\Gamma$', r'$X$', '', r'$M$', '', r'$Y$', '', r'$\Gamma$']

# 5️⃣ 求解能带结构
k_vec, k_dist, k_node = nso.model.k_path(k_path, 200)
evals = nso.model.solve_all(k_vec)

# 6️⃣ 画图
plt.figure(figsize=(8, 6))
for band in evals:
    plt.plot(k_dist, band, lw=1)

for node in k_node:
    plt.axvline(x=node, color='gray', lw=0.5, linestyle='--')

plt.xticks(k_node, k_labels)
plt.xlabel("K-path")
plt.ylabel("Energy (eV)")
plt.title("Band Structure from wannier90_hr.dat")
plt.grid(True)
plt.tight_layout()
plt.savefig("band_structure.png", dpi=300)
plt.show()
```
